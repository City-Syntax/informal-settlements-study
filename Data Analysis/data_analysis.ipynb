# Research Project: Cooling Interventions for Informal Settlements in Bihar, India

## Project Overview

This research project focuses on studying cooling interventions for informal settlements and slums in Bihar, India, particularly for structures made of scrap materials. The primary objectives are:

1. Assess the effectiveness of two cooling interventions: Radiant Barrier Foils (RBFs) and Mylar Emergency Blankets (MEBs)
2. Collect temperature and humidity data using data loggers in various houses across different settlements
3. Compare the effectiveness of interventions in shaded vs. non-shaded structures
4. Analyze the impact of these interventions on indoor temperatures during hot summer conditions

## Methodology

1. Place data loggers in multiple settlements:
   - Rainbow Field Settlements
   - Sports Complex Settlements
   - Control group near a Petrol Pump
2. Collect baseline data for 15 days before implementing interventions
3. Implement RBFs and MEBs in different houses
4. Collect post-intervention data to compare with baseline readings

The project aims to provide insights into effective, low-cost cooling solutions for informal settlements in hot climates, potentially improving living conditions for residents in these areas.

### This following script processes logger data files, detecting their encoding and cleans the logger data.

It includes functions to:
- Detect file encoding to handle potential Unicode errors.
- Clean and reformat U logger data.
- Clean and reformat R logger data.
- Process all logger files in a specified directory, clean them, and save the cleaned data to a new folder.

````python
import os
import pandas as pd
import chardet

def detect_encoding(file_path):
    """Detect the file encoding to handle potential unicode errors."""
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

def clean_u_logger_data(file_path):
    """Clean and reformat U logger data."""
    encoding = detect_encoding(file_path)
    df = pd.read_csv(file_path, encoding=encoding)

    df['Date'] = pd.to_datetime(df['Date'], format="%b %d, %Y %I:%M %p")
    df['Time'] = df['Date'].dt.strftime('%H:%M')
    df['Date'] = df['Date'].dt.date

    columns = ['Date', 'Time'] + [col for col in df.columns if col not in ['Date', 'Time']]
    df = df[columns]

    return df

def clean_r_logger_data(file_path):
    """Clean and reformat R logger data."""
    encoding = detect_encoding(file_path)
    df = pd.read_csv(file_path, sep='\t', encoding=encoding, skiprows=1)

    df['Time'] = pd.to_datetime(df['Time'], format="%Y-%m-%d %H:%M:%S")
    df['Date'] = df['Time'].dt.date
    df['Time'] = df['Time'].dt.strftime('%H:%M')

    columns = ['Date', 'Time', 'Temperature(C)', 'Humidity(%RH)']
    df = df[columns]

    return df

def process_loggers(data_dir):
    """Process all loggers and save cleaned data to a new folder."""
    u_loggers = []
    r_loggers = []

    cleaned_data_dir = os.path.join(os.path.dirname(data_dir), 'Cleaned_Data')
    os.makedirs(cleaned_data_dir, exist_ok=True)

    for file_name in os.listdir(data_dir):
        if file_name.endswith('_data.csv'):
            logger_id = file_name.split('_')[0]
            file_path = os.path.join(data_dir, file_name)
            cleaned_file_path = os.path.join(cleaned_data_dir, file_name)

            if file_name.startswith('U-'):
                u_loggers.append(logger_id)
                cleaned_df = clean_u_logger_data(file_path)
            elif file_name.startswith('R-'):
                r_loggers.append(logger_id)
                cleaned_df = clean_r_logger_data(file_path)
            else:
                continue

            cleaned_df.to_csv(cleaned_file_path, index=False, encoding='utf-8-sig')
    
    return u_loggers, r_loggers

if __name__ == "__main__":
    data_dir = './Loggers Data'
    u_loggers, r_loggers = process_loggers(data_dir)
    
    print("U Loggers available:", u_loggers)
    print("R Loggers available:", r_loggers)
````

### This following script reformats the environment data file.

It includes functions to:
- Clean the date and time columns in a more readable format.

````python
import pandas as pd
import os

def clean_environmental_data(file_path):
    """Clean and reformat Environmental Data."""
    df = pd.read_csv(file_path)

    df['datetime'] = pd.to_datetime(df['datetime'])

    df['Date'] = df['datetime'].dt.date
    df['Time'] = df['datetime'].dt.strftime('%H:%M')

    df = df.drop(columns=['datetime'])

    columns = ['Date', 'Time'] + [col for col in df.columns if col not in ['Date', 'Time']]
    df = df[columns]

    return df

if __name__ == "__main__":
    base_dir = os.path.dirname(os.path.abspath(__file__))
    file_path = os.path.join(base_dir, 'Environmental Data.csv')
    
    cleaned_df = clean_environmental_data(file_path)
    
    cleaned_df.to_csv(file_path, index=False)
    
    print("Environmental Data cleaned and saved successfully.")
````

### This following script combines the temperature info from the environmental data and all the logger data files.

1. **Reads logger data** from CSV files, combining 'Date' and 'Time' columns into a 'DateTime' index.
2. **Aggregates logger data** into a master DataFrame with a one-minute frequency.
3. **Reads and processes environmental data**, resampling and interpolating it to match the logger data frequency.
4. **Combines logger and environmental data** into the master DataFrame.
5. **Saves the master DataFrame** to a CSV file.
6. **Prints summary information** including the first few rows, shape, NaN counts, and first non-NaN entries for each logger.

Produces `master_dataframe.csv` file:

````python
import pandas as pd
import numpy as np
import os
import glob
import chardet

def detect_encoding(file_path):
    with open(file_path, 'rb') as file:
        raw_data = file.read()
    return chardet.detect(raw_data)['encoding']

def read_logger_data(file_path):
    encoding = detect_encoding(file_path)
    df = pd.read_csv(file_path, encoding=encoding)
    df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
    return df.set_index('DateTime')[['Temperature_Celsius(℃)']] if 'Temperature_Celsius(℃)' in df.columns else df.set_index('DateTime')[['Temperature(C)']]

logger_files = glob.glob('Cleaned Data/*.csv')
logger_dfs = {}

for file in logger_files:
    logger_id = os.path.basename(file).split('_')[0]
    logger_dfs[logger_id] = read_logger_data(file)

start_date = min(df.index.min() for df in logger_dfs.values())
end_date = max(df.index.max() for df in logger_dfs.values())
master_index = pd.date_range(start=start_date, end=end_date, freq='T')

master_df = pd.DataFrame(index=master_index)

for logger_id, df in logger_dfs.items():
    master_df[f'{logger_id}'] = df.reindex(master_index)

env_data = pd.read_csv('Environmental Data.csv')
env_data['DateTime'] = pd.to_datetime(env_data['Date'] + ' ' + env_data['Time'])
env_data = env_data.set_index('DateTime')

env_data_resampled = env_data.resample('T').interpolate()

master_df['Env_Temperature'] = env_data_resampled['temp']

master_df.to_csv('master_dataframe.csv')

print("Master dataframe created and saved as 'master_dataframe.csv'")
print(master_df.head())
print(f"Shape of master dataframe: {master_df.shape}")

nan_counts = master_df.isna().sum()
print("\nNumber of NaN values in each column:")
print(nan_counts)

print("\nFirst non-NaN entry for each logger:")
for column in master_df.columns:
    first_valid = master_df[column].first_valid_index()
    if first_valid is not None:
        print(f"{column}: {first_valid}")
    else:
        print(f"{column}: All values are NaN")
````

### Subtraction of environmental data from logger data to account for weather variations.

New columns are added to the master DataFrame for each logger, containing the difference between the logger temperature and the environmental temperature. This subtraction helps to account for weather variations and provides a clearer picture of the indoor temperature changes. The columns would include:

- Temperature difference (Logger temp - Environmental temp)
- Time period flag (Baseline, Transition, Post-intervention)

Produces `temperature_differences.csv` and `logger_flags.csv` files: 

````python
import pandas as pd
import numpy as np

df = pd.read_csv('master_dataframe.csv', index_col='DateTime', parse_dates=True)

logger_columns = [col for col in df.columns if col != 'Env_Temperature']

diff_df = pd.DataFrame(index=df.index)

for col in logger_columns:
    diff_df[col] = np.where(df[col].notna(), df[col] - df['Env_Temperature'], np.nan)

diff_df.to_csv('temperature_differences.csv')

print("Temperature differences saved to 'temperature_differences.csv'")
````

      - Analyze if the effectiveness of interventions differs between the two settlements.

      - Analyze how interventions perform during dry vs. rainy periods.

6. **Statistical Testing and Modeling:**

      - Conduct t-tests or ANOVA to compare mean temperatures between intervention groups.
      - Perform paired t-tests to compare before and after temperatures for each intervention.

   b) Regression analysis:

### This script will calculate the daily averages and standard deviations for both day and night separately, using the temperature differences data.

- **Day/Night Segmentation:** The data is split into day and night periods based on specific time ranges (6 AM to 7 PM for the time period during summer).
- **Statistical Calculations:** For each day, the average and standard deviation of temperature differences are computed separately for day and night.
- **Rain Detection:** Rain data from the environmental file is used to flag whether it rained on each day.
- **CSV Output:** The final results, including calculated statistics and rain data, are saved in a new CSV file.

Produces `daily_temperature_metrics.csv` file:

````python
import pandas as pd
import numpy as np

temperature_differences_df = pd.read_csv('temperature_differences.csv', parse_dates=['DateTime'])
environmental_data_df = pd.read_csv('Environmental Data.csv', parse_dates=[['Date', 'Time']])

day_start = '06:00:00'
day_end = '19:00:00'

# Function to check if it rained during the day
def did_it_rain(date):
    rain_data = environmental_data_df[
        (environmental_data_df['Date_Time'].dt.date == date) & 
        (environmental_data_df['preciptype'].notna())
    ]
    return not rain_data.empty

data = []

for date in temperature_differences_df['DateTime'].dt.date.unique():
    day_data = {}
    
    day_mask = (temperature_differences_df['DateTime'].dt.date == date) & \
               (temperature_differences_df['DateTime'].dt.time >= pd.to_datetime(day_start).time()) & \
               (temperature_differences_df['DateTime'].dt.time < pd.to_datetime(day_end).time())
               
    night_mask = (temperature_differences_df['DateTime'].dt.date == date) & \
                 ((temperature_differences_df['DateTime'].dt.time < pd.to_datetime(day_start).time()) | \
                 (temperature_differences_df['DateTime'].dt.time >= pd.to_datetime(day_end).time()))
    
    day_temps = temperature_differences_df[day_mask]
    night_temps = temperature_differences_df[night_mask]
    
    for logger in temperature_differences_df.columns[1:]:
        day_temps_logger = day_temps[logger].dropna()
        night_temps_logger = night_temps[logger].dropna()
        
        day_data[f'{logger}_day_avg'] = day_temps_logger.mean()
        day_data[f'{logger}_day_std'] = day_temps_logger.std()
        day_data[f'{logger}_day_min'] = day_temps_logger.min()
        day_data[f'{logger}_day_max'] = day_temps_logger.max()
        
        day_data[f'{logger}_night_avg'] = night_temps_logger.mean()
        day_data[f'{logger}_night_std'] = night_temps_logger.std()
        day_data[f'{logger}_night_min'] = night_temps_logger.min()
        day_data[f'{logger}_night_max'] = night_temps_logger.max()
    
    # Determine if it rained that day
    day_data['Date'] = date
    day_data['Rain'] = did_it_rain(date)
    
    data.append(day_data)

daily_metrics_df = pd.DataFrame(data)

final_columns = ['Date', 'Rain'] + [col for col in daily_metrics_df.columns if col not in ['Date', 'Rain']]
daily_metrics_df = daily_metrics_df[final_columns]

daily_metrics_df.to_csv('daily_temperature_metrics.csv', index=False)
````

### Statistical analysis for the intervention in any settlement. 

This script performs several statistical analyses to evaluate the impact of an intervention in a specific settlement. The analyses include:

1. **Wilcoxon Signed-Rank Test:** Assesses if there is a significant difference in temperature reductions between the pre- and post-intervention periods. A non-parametric test to determine if the median temperature reductions are significantly different.
2. **Effect Size Measurements:**
   - **Cohen's d:** Measures the standardized mean difference between pre- and post-intervention temperatures.
   - **Cliff's Delta:** A non-parametric effect size measure.
3. **OLS Regression Analysis:** Evaluates the relationship between the intervention period and temperature differences.
4. **ANOVA:** Tests for significant differences in temperature reductions.
5. **P-value Hypothesis Test**: To test whether the temperature reduction after the intervention is statistically significant compared to the control.

### **Hypothesis for the P-value Test:**

- **Null Hypothesis (H0)**: The intervention does not result in a statistically significant reduction in temperature compared to the control (no intervention).
- **Alternative Hypothesis (H1)**: The intervention results in a statistically significant reduction in temperature compared to the control.

````python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import wilcoxon
from scipy import stats

temperature_differences_df = pd.read_csv('temperature_differences.csv', parse_dates=['DateTime'])
logger_flags_df = pd.read_csv('logger_flags.csv')

settlement = 'Sports Complex'
intervention_type = 'RBF'
shaded = False

loggers_subset = logger_flags_df[
                                (logger_flags_df['Settlement'] == settlement) 
                                & (logger_flags_df['Intervention'] == intervention_type) 
                                # Uncomment the line below to filter by shaded/unshaded loggers
                                # & (logger_flags_df['Shaded'] == shaded)
                                ]

pre_intervention_data = []
post_intervention_data = []

for _, row in loggers_subset.iterrows():
    logger = row['Loggers']
    intervention_start = pd.to_datetime(row['Intervention_Start'])
    
    # Pre-intervention data (before intervention)
    pre_data = temperature_differences_df[(temperature_differences_df['DateTime'] < intervention_start)][logger]
    pre_intervention_data.extend(pre_data.dropna().tolist())

    # Post-intervention data (after intervention)
    post_data = temperature_differences_df[(temperature_differences_df['DateTime'] >= intervention_start)][logger]
    post_intervention_data.extend(post_data.dropna().tolist())

# Ensure both datasets are the same length
min_length = min(len(pre_intervention_data), len(post_intervention_data))
pre_intervention_data = pre_intervention_data[:min_length]
post_intervention_data = post_intervention_data[:min_length]

pre_data_series = np.array(pre_intervention_data)
post_data_series = np.array(post_intervention_data)

# Wilcoxon signed-rank test
w_statistic, p_value = wilcoxon(post_data_series, pre_data_series)
print(f"Wilcoxon signed-rank test statistic: {w_statistic}, p-value: {p_value}")

# Hypothesis for p-value test
# Null hypothesis (H0): There is no significant difference in temperature reductions.
# Alternative hypothesis (H1): There is a significant difference in temperature reductions.
alpha = 0.05
if p_value < alpha:
    print("\nNull hypothesis rejected: There is a significant difference in temperature reductions.")
else:
    print("\nNull hypothesis accepted: There is no significant difference in temperature reductions.")

mean_diff = np.mean(post_data_series - pre_data_series)
std_diff = np.std(post_data_series - pre_data_series, ddof=1)

# Calculate Cliff's delta
def cliff_delta(a, b):
    """
    Calculate Cliff's Delta for non-parametric effect size.
    """
    a = np.array(a)
    b = np.array(b)
    n1 = len(a)
    n2 = len(b)
    combined = np.concatenate([a, b])
    n1_over_n2 = 0
    for i in range(n1):
        n1_over_n2 += np.sum(combined[i] > combined[n1:])
    return (n1_over_n2 - (n1 * n2 / 2)) / (n1 * n2)

cliffs_delta = cliff_delta(pre_data_series, post_data_series)
print(f"\nCliff's delta: {cliffs_delta}")

# OLS Regression
df = pd.DataFrame({
    'TemperatureDifference': np.concatenate([pre_data_series, post_data_series]),
    'InterventionPeriod': ['Pre']*len(pre_data_series) + ['Post']*len(post_data_series)
})

model = ols('TemperatureDifference ~ C(InterventionPeriod)', data=df).fit()

print("\n\nOLS Regression Results:")
print(model.summary())

# Perform ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)
print("\n\nANOVA Results:")
print(anova_table)
````

````python
import pandas as pd
import numpy as np

def calculate_temperature_reduction(settlement, intervention_type, shaded=None):
    """
    Calculates the average temperature reduction caused by a specific intervention in a specific settlement.
    
    Parameters:
    - settlement (str): The settlement name to filter the data.
    - intervention_type (str): The type of intervention (RBF or MEB).
    - shaded (bool, optional): Whether to filter by shaded loggers (True) or unshaded (False). 
      If None, do not filter by shading.
      
    Returns:
    - float: The average temperature reduction in degrees Celsius.
    """
    temperature_differences_df = pd.read_csv('temperature_differences.csv', parse_dates=['DateTime'])
    logger_flags_df = pd.read_csv('logger_flags.csv')

    # Filter loggers for the given settlement and intervention type
    loggers_subset = logger_flags_df[
        (logger_flags_df['Settlement'] == settlement) 
        & (logger_flags_df['Intervention'] == intervention_type)
        # Uncomment the line below to filter by shaded/unshaded loggers
        # & (logger_flags_df['Shaded'] == shaded)
    ]
    
    if shaded is not None:
        loggers_subset = loggers_subset[loggers_subset['Shaded'] == shaded]

    pre_intervention_data = []
    post_intervention_data = []

    for _, row in loggers_subset.iterrows():
        logger = row['Loggers']
        intervention_start = pd.to_datetime(row['Intervention_Start'])
        
        pre_data = temperature_differences_df[(temperature_differences_df['DateTime'] < intervention_start)][logger]
        pre_intervention_data.extend(pre_data.dropna().tolist())

        post_data = temperature_differences_df[(temperature_differences_df['DateTime'] >= intervention_start)][logger]
        post_intervention_data.extend(post_data.dropna().tolist())

    min_length = min(len(pre_intervention_data), len(post_intervention_data))
    pre_intervention_data = pre_intervention_data[:min_length]
    post_intervention_data = post_intervention_data[:min_length]

    pre_data_series = np.array(pre_intervention_data)
    post_data_series = np.array(post_intervention_data)

    avg_pre_temperature = np.mean(pre_data_series)
    avg_post_temperature = np.mean(post_data_series)
    
    avg_temperature_reduction = avg_pre_temperature - avg_post_temperature
    
    return avg_temperature_reduction

settlement = 'Sports Complex'
intervention_type = 'RBF'
shaded = True
average_reduction = calculate_temperature_reduction(settlement, intervention_type, shaded=False)
print(f"Average temperature reduction for {intervention_type} in {settlement}: {average_reduction:.2f}°C")
````
